# Exploratory Data Analysis (EDA)

## Objective

The objective of this exploratory data analysis is to understand the structure and quality of the IBM Telco Customer Churn dataset before building machine learning models.

This analysis focuses on:

- Understanding the dataset dimensions
- Inspecting data types
- Detecting missing values
- Identifying potential data quality issues
- Exploring feature distributions
- Investigating relationships between customer attributes and churn
- Discovering useful insights for feature engineering and model development

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_excel("Telco_customer_churn.xlsx")
```

## Data Types

The dataset contains a mixture of numerical and categorical features.

- Numerical variables include customer tenure, monthly charges, total charges, latitude, longitude, churn score, and CLTV.
- Categorical variables describe customer demographics, subscribed services, contract information, payment methods, and churn status.

The combination of numerical and categorical variables makes the dataset suitable for both statistical analysis and machine learning.

```python
identifier = [
    'CustomerID'
]

geographic = [
    'Country',
    'State',
    'City',
    'Zip Code',
    'Lat Long',
    'Latitude',
    'Longitude'
]

demographic = [
    'Gender',
    'Senior Citizen',
    'Partner',
    'Dependents'
]

service = [
    'Phone Service',
    'Multiple Lines',
    'Internet Service',
    'Online Security',
    'Online Backup',
    'Device Protection',
    'Tech Support',
    'Streaming TV',
    'Streaming Movies'
]

account = [
    'Tenure Months',
    'Contract',
    'Paperless Billing',
    'Payment Method',
    'Monthly Charges',
    'Total Charges'
]

target_related = [
    'Churn Label',
    'Churn Value',
    'Churn Score',
    'CLTV',
    'Churn Category',
    'Churn Reason'
]
categories = {
    "Identifier": identifier,
    "Geographic": geographic,
    "Demographic": demographic,
    "Service": service,
    "Account": account,
    "Target-related": target_related
}

for category, cols in categories.items():
    print(f"{category}: {len(cols)} features")
```

## Target Variable Analysis

The target variable is **Churn Label**, which indicates whether a customer has left the company's services.

Understanding the distribution of the target variable is essential before training machine learning models, as it helps identify potential class imbalance and provides an overview of customer retention behavior.

```python
churn_counts = df['Churn Label'].value_counts()


churn_percent = (
    df['Churn Label']
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
)

display(
    pd.DataFrame({
        "Count": churn_counts,
        "Percentage (%)": churn_percent
    })
)
```

```python
sns.set_theme(style="whitegrid")

fig, ax = plt.subplots(figsize=(6,4))

palette = ["#4C72B0", "#DD8452"]

sns.countplot(
    data=df,
    x="Churn Label",
    palette=palette,
    edgecolor="black",
    linewidth=1.2,
    ax=ax
)

total = len(df)

for p in ax.patches:
    count = int(p.get_height())
    percentage = 100 * count / total

    ax.annotate(
        f"{count:,}\n({percentage:.1f}%)",
        (p.get_x() + p.get_width()/2, count),
        ha="center",
        va="bottom",
        fontsize=11,
        fontweight="bold"
    )

ax.set_title(
    "Customer Churn Distribution",
    fontsize=18,
    weight="bold",
    pad=20
)

ax.set_xlabel("")
ax.set_ylabel("Number of Customers", fontsize=12)

sns.despine()

plt.tight_layout()
plt.show()
```

### Insights

- The dataset is moderately imbalanced, with significantly more non-churn customers than churn customers.
- Approximately three-quarters of customers remained with the company, while about one-quarter discontinued the service.
- Although the imbalance is not severe, it should be considered during model development and evaluation.
- Performance metrics such as Precision, Recall, F1-score, and ROC-AUC should be prioritized over Accuracy when comparing classification models.

```python
contract_churn = pd.crosstab(
    df["Contract"],
    df["Churn Label"]
)

contract_churn
```

```python
contract_pct = pd.crosstab(
    df["Contract"],
    df["Churn Label"],
    normalize="index"
).round(3) * 100

contract_pct
```

```python
contract_pct.plot(
    kind="bar",
    stacked=True,
    figsize=(7,4),
    color=["#4C72B0", "#DD8452"],
    edgecolor="black"
)

plt.title(
    "Customer Churn Rate by Contract Type",
    fontsize=18,
    weight="bold"
)

plt.xlabel("")
plt.ylabel("Percentage (%)")

plt.xticks(rotation=0)

plt.legend(
    title="Churn",
    bbox_to_anchor=(1.02,1),
    loc="upper left"
)

plt.grid(axis="y", alpha=.25)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(7,4))

sns.barplot(
    data=df,
    x="Contract",
    y="Churn Value",
    estimator=np.mean,
    palette="viridis"
)

plt.ylabel("Average Churn Rate")
plt.xlabel("")
plt.title(
    "Average Churn Rate by Contract Type",
    fontsize=18,
    weight="bold"
)

plt.ylim(0,0.6)

for container in plt.gca().containers:
    plt.bar_label(
        container,
        fmt="%.2f",
        padding=3,
        fontsize=11
    )

sns.despine()

plt.tight_layout()

plt.show()
```

### Insights
- Customers with month-to-month contracts exhibit the highest churn rate.
- The churn rate decreases substantially for customers with one-year contracts.
- Customers with two-year contracts demonstrate the lowest likelihood of churning.
- These findings suggest that longer contract commitments are strongly associated with improved customer retention.
- Contract type is expected to be one of the most influential predictors in the churn prediction model.

## Customer Tenure Analysis

Customer tenure represents the number of months a customer has remained with the company.

This feature is expected to have a strong relationship with customer churn, as long-term customers are generally more loyal than newly acquired customers.

The following analysis explores:

- Distribution of customer tenure
- Comparison of tenure between churned and retained customers
- Churn rate across different tenure groups

```python
plt.figure(figsize=(8,4))

sns.histplot(
    data=df,
    x="Tenure Months",
    bins=30,
    kde=True,
    color="#4C72B0",
    edgecolor="white",
    linewidth=.8
)

plt.title(
    "Distribution of Customer Tenure",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Tenure (Months)")
plt.ylabel("Number of Customers")

sns.despine()

plt.tight_layout()
plt.show()
```

### Insight

The tenure distribution provides an overview of customer longevity.

A concentration of customers with relatively short tenure may indicate continuous acquisition of new customers, while a higher proportion of long-tenure customers suggests stronger customer retention.

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    data=df,
    x="Churn Label",
    y="Tenure Months",
    palette=["#4C72B0","#DD8452"],
    width=.5,
    linewidth=1.5
)

plt.title(
    "Customer Tenure by Churn Status",
    fontsize=18,
    weight="bold"
)

plt.xlabel("")
plt.ylabel("Tenure (Months)")

sns.despine()

plt.tight_layout()

plt.show()
```

### Insight

Customers who churn generally have shorter tenure compared to retained customers.

This suggests that customer loyalty increases with the length of the relationship, making tenure a potentially important predictive feature.

```python
bins = [0,12,24,36,48,60,72]

labels = [
    "0-12",
    "13-24",
    "25-36",
    "37-48",
    "49-60",
    "61-72"
]

df["Tenure Group"] = pd.cut(
    df["Tenure Months"],
    bins=bins,
    labels=labels,
    include_lowest=True
)
tenure_churn = (
    df.groupby("Tenure Group")["Churn Value"]
      .mean()
      .mul(100)
      .reset_index()
)

tenure_churn
```

```python
plt.figure(figsize=(8,6))

ax = sns.barplot(
    data=tenure_churn,
    x="Tenure Group",
    y="Churn Value",
    palette="Blues_r"
)

for container in ax.containers:
    ax.bar_label(
        container,
        fmt="%.1f%%",
        fontsize=11,
        padding=3
    )

plt.title(
    "Customer Churn Rate Across Tenure Groups",
    fontsize=18,
    weight="bold"
)

plt.xlabel("Tenure (Months)")
plt.ylabel("Churn Rate (%)")

plt.ylim(0,50)

sns.despine()

plt.tight_layout()

plt.show()
```

### Insights

- Customers with less than one year of tenure exhibit the highest churn rate.
- The likelihood of churn decreases steadily as customer tenure increases.
- Long-term customers are significantly more likely to remain with the company.
- These findings indicate that customer retention initiatives should focus primarily on newly acquired customers during their first year.

## Internet Service vs Customer Churn

Internet service type may influence customer satisfaction and retention.

This analysis compares churn behavior among customers with different internet service types to identify whether certain services are associated with higher churn rates.

```python
plt.figure(figsize=(6,4))

ax = sns.countplot(
    data=df,
    x="Internet Service",
    palette="Set2",
    edgecolor="black"
)

total = len(df)

for p in ax.patches:
    count = int(p.get_height())
    percentage = count / total * 100

    ax.annotate(
        f"{percentage:.1f}%",
        (p.get_x() + p.get_width()/2, count),
        ha="center",
        va="bottom",
        fontsize=11,
        weight="bold"
    )

plt.title(
    "Distribution of Internet Service Types",
    fontsize=17,
    weight="bold"
)

plt.xlabel("")
plt.ylabel("Number of Customers")

sns.despine()

plt.tight_layout()
plt.show()
```

```python
internet_churn = (
    df.groupby("Internet Service")["Churn Value"]
      .mean()
      .mul(100)
      .reset_index()
      .sort_values("Churn Value", ascending=False)
)

internet_churn
```

```python
plt.figure(figsize=(6,4))

ax = sns.barplot(
    data=internet_churn,
    x="Internet Service",
    y="Churn Value",
    palette=["#DD8452", "#55A868", "#4C72B0"]
)

for container in ax.containers:
    ax.bar_label(
        container,
        fmt="%.1f%%",
        fontsize=11,
        padding=3
    )

plt.title(
    "Churn Rate by Internet Service",
    fontsize=17,
    weight="bold"
)

plt.xlabel("")
plt.ylabel("Churn Rate (%)")

sns.despine()

plt.tight_layout()

plt.show()
```

### Insights

- Churn rates vary considerably across internet service types.
- Customers using Fiber Optic service exhibit the highest churn rate.
- DSL customers show a lower churn rate, while customers without internet service have the lowest likelihood of churning.
- These findings suggest that internet service type is an important factor influencing customer retention.

## Correlation Analysis

Correlation analysis helps identify linear relationships between numerical features in the dataset.

The heatmap below illustrates the Pearson correlation coefficients among the numerical variables. Strong positive correlations are shown in darker red, while strong negative correlations appear in darker blue.

```python
numeric_cols = df.select_dtypes(include="number")

numeric_cols = numeric_cols.loc[:, numeric_cols.nunique() > 1]
```

```python
corr_matrix = numeric_cols.corr()

corr_matrix.round(2)
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    square=True,
    cbar_kws={"shrink": 0.8}
)

plt.title(
    "Correlation Matrix of Numerical Features",
    fontsize=17,
    weight="bold",
    pad=15
)

plt.tight_layout()
plt.show()
```

### Insights

- Most numerical features exhibit weak to moderate correlations, indicating limited multicollinearity.
- **Tenure Months** and **Total Charges** show a strong positive correlation, as customers with longer service duration generally accumulate higher total charges.
- **Monthly Charges** has a moderate positive relationship with **Total Charges**.
- **Churn Value** is negatively correlated with **Tenure Months**, suggesting that long-term customers are less likely to churn.
- Features such as **Churn Score** are highly correlated with the target because they are derived from churn-related information and should be excluded during model training to prevent data leakage.
